```python
import pandas as pd
ds = pd.read_csv('Students.csv')
ds.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 
heatmap_data = ds.pivot_table(
    index= 'Zone_EN',
    columns = 'Gender_EN',
    values='Student_Count',
    aggfunc='sum'
)
plt.figure(figsize=(10,6))
sns.heatmap(heatmap_data, annot=True, fmt=".0f", cmap="Blues",cbar_kws={'label': 'Number of Students'})
plt.show()

    
```
